```python
import random

# Read the u.data file
original_data = []

with open('our data/u.data', 'r') as file:
    for line in file:
        user_id, item_id = map(int, line.strip().split('\t')[:2])
        original_data.append((user_id, item_id))

# Collect unique user IDs and item IDs
user_ids = set()
item_ids = set()

for user_id, item_id in original_data:
    user_ids.add(user_id)
    item_ids.add(item_id)

# Create negative samples
negative_samples = []

print("Generating negative samples...")
for user_id in user_ids:
    for _ in range(99):  # Generate 25 negative samples for each user
        while True:
            item_id = random.choice(list(item_ids))  # Randomly select an item ID

            # Check if the user has interacted with the item
            if (user_id, item_id) not in original_data:
                negative_samples.append((user_id, item_id))
                break
                   
```

```python
fd = open('our data/u.data.negative', 'w')
for user_id in user_ids:
    while True:
        item_id = random.choice(list(item_ids))  # Randomly select an item ID
        
        if len([x for x in negative_samples if x[1] == item_id]) >= 99:
            continue

        # Check if item_id have 99 negative samples
        if [user_id, item_id] in negative_samples:
            continue

        print(user_id, item_id)
        fd.write("("+str(user_id) + ',' + str(item_id) + ")\t")

        i = 0
        for user, item in negative_samples:
          if item == item_id and user != user_id:
            fd.write(str(user)+"\t")
            i=+1
            if i == 99:
              break

        fd.write("\n")
        break
fd.close()
```

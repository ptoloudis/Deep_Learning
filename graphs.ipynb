```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data 
top_k = pd.read_csv('our data/top_k.csv')
print(top_k.head())
```

```python
# plot the data in axis x is 1 to 10 and y is the number of items in top k
x = range(1, 11)

# Plot the data
plt.plot(x, top_k['HR'].values)
plt.xlabel('Number of items in top k')
plt.ylabel('HR@K')
plt.title('HR@K vs Number of items in top k')

# Show the plot
plt.show()
```

```python
plt.plot(x, top_k['NDCG'].values)
plt.xlabel('Number of items in top k')
plt.ylabel('NDCG@K')
plt.title('NDCG@K vs Number of items in top k')

# Adjust the layout of the subplots
plt.show()
```

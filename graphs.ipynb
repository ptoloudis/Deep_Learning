```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data 
top_k = pd.read_csv('our data/top_k.csv')
print(top_k.head())
```

```python
# plot the data in axis x is 1 to 10 and y is the number of items in top k
x = range(1, 11)

# Plot the data
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x, top_k['HR'].values)
ax1.set_xlabel('Number of items in top k')
ax1.set_ylabel('HR@K')
ax1.set_title('HR@K vs Number of items in top k')

ax2.plot(x, top_k['NDCG'].values)
ax2.set_xlabel('Number of items in top k')
ax2.set_ylabel('NDCG@K')
ax2.set_title('NDCG@K vs Number of items in top k')

# Adjust the layout of the subplots
plt.tight_layout()

# Show the plot
plt.show()
```
